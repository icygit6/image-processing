```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, morphology, color
from skimage.filters import threshold_otsu
import time
path = "/content/drive/MyDrive/python/IMAGE PROCESSING/lena.png"
image = io.imread(path)
gray = color.rgb2gray(image)
thresh = threshold_otsu(gray)
bin = gray > thresh #creates a binary image by separating the foreground (objects) from the background.
start = time.time()
skeleton = morphology.skeletonize(bin) #reduces objects in the binary image to their skeletal representations while preserving connectivity.
exe = time.time() - start
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(bin, cmap="gray")
axes[0].set_title("Binary Image")
axes[0].axis("off")
axes[1].imshow(skeleton, cmap="gray")
axes[1].set_title("Skeletonized Image")
axes[1].axis("off")
plt.tight_layout()
plt.show()
print(f"Skeletonization Execution Time: {exe:.4f} seconds")
```
