```python
from skimage import io, exposure

# 讀取影像
image = io.imread('/content/drive/MyDrive/digImage/Unequalized_pic.jpg', as_gray=True)

# 對影像進行全域直方圖等化
equalized_image = exposure.equalize_hist(image)

io.imshow(equalized_image)
io.show()
```

```python
from skimage import io, exposure,util
from skimage.color import rgb2gray
from skimage.morphology import disk
from skimage.filters import rank

# 讀取影像
image = io.imread('/content/drive/MyDrive/digImage/Unequalized_pic.jpg', as_gray=True)
#把圖像轉成unit8
image = util.img_as_ubyte(image)

# 使用局部直方圖等化
image_equalize_30 = rank.equalize(image, footprint=disk(30))

# 顯示原始影像和處理後的影像
io.imshow(image_equalize_30, cmap='gray')
io.show()
```

```python
# 使用局部直方圖等化
image_equalize_50 = rank.equalize(image, footprint=disk(50))

# 顯示原始影像和處理後的影像
io.imshow(image_equalize_50, cmap='gray')
io.show()
```

```python
from PIL import Image,ImageOps
import matplotlib.pyplot as plt
import numpy as np

image = Image.open('/content/drive/MyDrive/digImage/Unequalized_pic.jpg')
image = image.convert("L")
equalized_image = ImageOps.equalize(image)
equalized_image = np.asarray(equalized_image)
plt.imshow(equalized_image,cmap='gray')
plt.show()
```
