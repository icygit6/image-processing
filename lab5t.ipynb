```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def unwarp(path, src, dst, out=(500, 500)):
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    height, width, _ = image.shape
    M = cv2.getPerspectiveTransform(src, dst)
    warp = cv2.warpPerspective(image, M, out, flags=cv2.INTER_LINEAR)
    gray = cv2.cvtColor(warp, cv2.COLOR_RGB2GRAY)
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title("Original Image")
    plt.title(f"Original Image\nWidth: {width}, Height: {height}")
    plt.scatter(src[:, 0], src[:, 1], color="red")
    plt.subplot(1, 2, 2)
    plt.imshow(gray, cmap="gray")
    plt.title("Warped Skull Region")
    plt.tight_layout()
    plt.show()
path = "/content/drive/MyDrive/python/IMAGE PROCESSING/ambassadors.jpg"

# Coordinates for the skull region (adjusted to fit the image size)
src = np.float32([#x,y
    (145, 430),  # Top-left corner of the skull region
    (395, 340),  # Top-right corner
    (295, 455),  # Bottom-right corner
    (75, 540)   # Bottom-left corner
])
# Destination points for the output (top-down view)
dst = np.float32([
    (0, 0),       # Top-left corner of the output
    (500, 0),     # Top-right corner
    (500, 500),   # Bottom-right corner
    (0, 500)      # Bottom-left corner
])
unwarp(path, src, dst)
```
