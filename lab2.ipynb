**Commonly Used Colormaps**

*viridis* = A perceptually uniform colormap that transitions from yellow to blue.

*plasma* = A colormap that transitions from purple to yellow.


*inferno* = A colormap that goes from dark purple to yellow-orange.


*magma* = A colormap that transitions from black to bright yellow.


*cividis* = A colormap designed to be colorblind-friendly; it moves from blue to yellow.

**Diverging Colormaps:**

*seismic* = Transitions from blue through white to red.


*coolwarm* = A blue-to-red colormap, good for showing positive and negative trends.


*RdBu* = Red to blue colormap, similar to 'seismic' but smoother.


**Sequential Colormaps:**
*Blues* = A sequential colormap that transitions from white to dark blue.

*Reds* = A sequential colormap that transitions from white to dark red.

*Greens* = A sequential colormap that transitions from white to dark green.

**Cyclic Colormaps:**

*twilight* = Cyclic colormap transitioning through grays, purples, and greens.

*hsv* = A colormap that represents hue in a circular color spectrum.

*hot* = Transitions from black to red, orange, and yellow, mimicking thermal images.

*cool* = A colormap that transitions from cyan to magenta.

*spring* = Transitions from magenta to yellow.

*summer* = Transitions from green to yellow.

*autumn* = Transitions from red to yellow.

*winter* = Transitions from blue to green.

*plasma* = olormap transitions from dark purple to bright yellow.

*bone* = A grayscale colormap with a soft bluish tint.

*copper*  = A colormap that gives a copper-like appearance with shades of brown.

```python
# @title Exercise 6-2
import cv2
import numpy as np
import matplotlib.pyplot as plt
path = "/content/lena.png"
image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
mean = np.mean(image)
_, binimg = cv2.threshold(image, mean, 255, cv2.THRESH_BINARY)
plt.imshow(binimg, cmap="gray")
plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
path = "/content/lena.png"
imop = cv2.imread(path)
grayimg = cv2.cvtColor(imop, cv2.COLOR_BGR2GRAY)
gimg = np.array(grayimg)
mean = np.mean(grayimg)
_, binimg = cv2.threshold(gimg, mean, 256, cv2.THRESH_BINARY)
plt.imshow(binimg, cmap="gray")
plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
path = "/content/lena.png"
imop = cv2.imread(path)
grayimg = cv2.cvtColor(imop, cv2.COLOR_BGR2GRAY)
gimg = np.array(grayimg)
_, binimg = cv2.threshold(gimg, 128, 256, cv2.THRESH_BINARY)
plt.imshow(binimg, cmap="gray")
plt.axis("off")
plt.show()
```
